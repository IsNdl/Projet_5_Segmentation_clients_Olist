## Clustering using KMeans

### Data pre-processing
For the clustering, at this step only RFM features and rewiew score will be used.

```python
df = pd.read_csv('data_for_clustering', index_col=0)
```

```python
df.head()
```

```python
data = df[['total_number_of_orders',
           'days_since_order', 'total_payment_sum',
           'review_score']]

# Making a copy, includind order_date, in prevision of maintenance
data_bis = data.copy(deep=True)
data_bis['order_date'] = df['order_date']

data = data.set_index(df['customer_unique_id'])
data_bis = data_bis.set_index(df['customer_unique_id'])
```

```python
data.head()
# Recency [1]
# Frequency [0]
# Monetary (value) [2]
# Satisfaction [3]
```

```python
data.dtypes
```

```python
data_bis.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = data.values
X = StandardScaler().fit_transform(X)
```

### Elbow method for KMeans clustering

```python
inertia = []
k_range = range(1, 10)
for k in k_range:
    model = KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)
```

```python
plt.plot(k_range, inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Model cost (Inertia)')
```

The most accurate number of clusters is not very clear. To determine the best value for k, the algorithm will be trained with 4, 5 and 6 clusters.

### Determination of the best number of clusters
#### Clustering With 6 clusters

##### K-Means clustering

```python
model_6 = KMeans(n_clusters=6, random_state=7)
# init = 'k-means ++' by default
model_6.fit(X)
y_pred_6 = model_6.predict(X)
```

##### Graphical representation

```python
# Interactive plot
# creating figure
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
# creating the plot
plot_6_clusters = ax.scatter(X[:, 0], X[:, 1], X[:, 2],
                             c=y_pred_6, alpha=0.5, cmap='viridis')
# Recency [1]
# Frequency [0]
# Monetary (value) [2]
# Satisfaction [3]

# setting title and labels
ax.set_title('Clustering using KMeans (k = 6)')
ax.set_xlabel('Frequency')
ax.set_ylabel('Recency')
ax.set_zlabel('Monetary value')
plt.legend()
# displaying the plot
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y_pred_6, alpha=0.5, cmap='viridis')
ax.scatter(model_6.cluster_centers_[:, 0],
           model_6.cluster_centers_[:, 1],
           model_6.cluster_centers_[:, 2], c='r')
ax.set_title('Clustering using KMeans (6 clusters)')
ax.set_xlabel('Frequency')
ax.set_ylabel('Recency')
ax.set_zlabel('Monetary value')
plt.legend()
plt.show()
```

##### Clusters information

```python
data_6 = data.copy()
data_6['label'] = model_6.labels_
```

```python
gb6 = data_6.groupby(model_6.labels_)
```

```python
gb6.mean()
```

##### Conclusion : 
6 clusters seems not very appropriate (3 big clusters and 3 small with one < 200 customers) even though tendencies are appearing.

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from tqdm import tqdm
epsilon_list = np.arange(0.1, 0.5, 0.1)  # List of epsilon values
sample_list = np.arange(5, 25, 5)  # List of min samples number


def GridSearchDBSCAN(epsilon_list, sample_list, X_std):

    # X_std is standard array or df
    silhouette = []
    point = []
    samples = []

    for i in tqdm(range(len(epsilon_list))):
        for k in tqdm(range(len(sample_list))):
            dbscan = DBSCAN(eps=epsilon_list[i], min_samples=sample_list[k])
            dbscan.fit(X_std)
            silh = metrics.silhouette_score(X_std, dbscan.labels_)
            point.append(silh)
            if silh >= max(point):
                silh_coef = silh
                sample = sample_list[k]
        silhouette.append(silh_coef)
        samples.append(sample)

    import matplotlib.pyplot as plt
    plt.plot(epsilon_list, silhouette)
    plt.title("Silhouette score")
    plt.xlabel('epsilon')
    plt.show

    return(epsilon_list, silhouette, samples)
```

```python
GridSearchDBSCAN(epsilon_list, sample_list, X)
```

```python
from sklearn.cluster import dbscan
y_pred = DBSCAN(eps=0.2, min_samples=15).fit_predict(X)
```
