## Creation of the global dataset

### Files to merge
Identification of useful features for join/merge:
- customers --> 'customer_id'
- orders --> 'order_id', 'customer_id'
- products --> 'product_id'
- items --> 'order_id', 'order_item_id', 'product_id', 'seller_id'
- payments --> 'order_id'
- reviews --> ''

No use of _geolocation_ & _sellers_ (clustering focusing on clients)

### Customers + orders: building _Frequency_ feature

```python
df = pd.merge(customers, orders, on='customer_id', how='inner')
```

```python
# Counting the number of orders per unique customer
total_number_of_orders = pd.DataFrame(df.groupby('customer_unique_id')['order_id'].count())

# Rename the column in 'total_number_of_orders'
total_number_of_orders.columns = ['total_number_of_orders']

# Merging with dataframe
df = pd.merge(df, total_number_of_orders, on='customer_unique_id', how='inner')
```

```python
df.head()
```

### Recency

```python
import datetime as dt
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
recency = pd.DataFrame(df.groupby('customer_unique_id')['order_purchase_timestamp'].max())
recency.columns = ['order_date']
df = pd.merge(df, recency, on='customer_unique_id', how='inner')
```

```python
# Converting into a date object, on a new column
df['order_date'] = pd.to_datetime(df['order_purchase_timestamp'],
                                  format='%Y-%m-%d')
df['order_date'] = df['order_date'].dt.date   # ('%Y-%m-%d')
```

```python
df.head()
```

```python
today = dt.date.today()  # date of code execution, format YYYY, MM, DD
```

```python
df['days_since_order'] = (today - df['order_date']).dt.days
# Calculating difference of days since the order was payed
```

```python
df.head()
```

### Adding payments: building _Monetary_ feature

```python
df = pd.merge(df, payments, on='order_id', how='inner')
```

```python
# Counting total amount paid per single customer
total_payment = pd.DataFrame(df.groupby('customer_unique_id')['payment_value'].sum())
total_payment.columns = ['total_payment_sum']  # Rename column
# Merging with dataframe
df = pd.merge(df, total_payment, on='customer_unique_id', how='inner')
```

```python
df.head()
```

### Adding items

```python
df = pd.merge(df, items, on='order_id', how='inner')
```

```python
df.head()
```

### Adding products

```python
df = pd.merge(df, products, on='product_id', how='inner')
```

### Adding reviews: _Satifsfaction_

```python
df = pd.merge(df, reviews, on='order_id', how='inner')
```
